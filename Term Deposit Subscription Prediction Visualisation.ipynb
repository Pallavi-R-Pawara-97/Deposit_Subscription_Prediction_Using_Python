#### Transform The Data :

##### Transform non-numerical into numerical column :

```python
from sklearn.preprocessing import LabelEncoder
for column in df.columns:
        if df[column].dtype == np.number :
            continue
        df[column] = LabelEncoder().fit_transform(df[column])
```
